```python
%matplotlib inline
#dependencies & setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os 

#files to load (remember to change these)
city_data_to_load = os.path.join('data', 'city_data.csv')
ride_data_to_load = os.path.join('data', 'ride_data.csv')

#read the city and ride data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

#combine the data into a single dataset
ride_city_data = pd.merge(ride_data, city_data, how='left', on='city')

#display the data table for preview
ride_city_data.head()
```

```python
#groupby type & city from merged data set
groupby_city_type = ride_city_data.groupby(["type"] + ["city"])

#X-Axis
total_rides = groupby_city_type["ride_id"].count()

rides_urban = total_rides["Urban"]
rides_rural = total_rides["Rural"]
rides_suburban = total_rides["Suburban"]

#Y-Axis
average = groupby_city_type["fare"].mean()

avg_urban = average["Urban"]
avg_rural = average["Rural"]
avg_suburban = average["Suburban"]

#build the scatter plots for each city type
urban_plot=plt.scatter(rides_urban,avg_urban, marker="o", edgecolor="black",
           s=total_drivers*20, alpha=0.75, linewidth=1, c="lightcoral", label="Urban")

rural_plot=plt.scatter(rides_rural,avg_rural, marker="o", edgecolor="black",
           s=total_drivers*20, alpha=0.75, linewidth=1, c="gold", label="Rural")

suburban_plot=plt.scatter(rides_suburban,avg_suburban, marker="o", edgecolor="black",
           s=total_drivers*20, alpha=0.75, linewidth=1, c="skyblue", label="Suburban")

#incorporate the other graph properties
plt.grid()

#set lower & upper limits for X-Axis
plt.xlim(0, 42)

#set lower & upper limits for Y-Axis
plt.ylim(18, 45)

#create a legend
plt.legend(title="City Types",loc='best')

#create labels for the scatter plot
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")

#incorporate a text label regarding circle size
total_drivers = groupby_city_type["driver_count"].count()

#save figure
plt.savefig("../Images/pyber_bubble_plot.png")
```

```python
#calculate type percents
sum_city_type = groupby_city_type["fare"].sum()

sum_urban_fares = sum_city_type["Urban"].sum()
sum_rural_fares = sum_city_type["Rural"].sum()
sum_suburban_fares = sum_city_type["Suburban"].sum()

total_fares = sum_urban + sum_rural + sum_suburban 

urban_fares = (sum_urban_fares/total_fares) * 100 
rural_fares = (sum_rural_fares/total_fares) * 100 
suburban_fares = (sum_suburban_fares/total_fares) * 100 

print(urban_fares)
print(rural_fares)
print(suburban_fares)

#build pie chart
labels = ["Urban", "Rural", "Suburban"]
sizes = [62.7, 6.8, 30.5]
colors = ["lightcoral", "gold", "skyblue"]
explode = (0, .10 , .10)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=260)

plt.axis("equal")

plt.title("% of Total Fares by City Type")

#save figure
plt.savefig("../Images/pyber_fares_by_city_type.png")
```

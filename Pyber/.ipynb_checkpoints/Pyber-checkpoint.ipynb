```python
%matplotlib inline
#dependencies & setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os 

#files to load (remember to change these)
city_data_to_load = os.path.join('data', 'city_data.csv')
ride_data_to_load = os.path.join('data', 'ride_data.csv')

#read the city and ride data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

#combine the data into a single dataset
ride_city_data = pd.merge(ride_data, city_data, how='left', on='city')

#display the data table for preview
ride_city_data.head()
```

```python
#groupby city
complete_summary = ride_city_data.groupby(["city"] + ["type"] + ["driver_count"])

#calculations
fare = complete_summary["fare"].mean()
riders = complete_summary["ride_id"].count()

#create dataframe
summary = pd.DataFrame({"Avg. Fare": fare, "Total Riders": riders})

#reset dataframe index without creating a new one
summary.reset_index(inplace=True)

#sort out data by city type
urban = summary.loc[summary["type"] == "Urban", :]
suburban = summary.loc[summary["type"] == "Suburban", :]
rural = summary.loc[summary["type"] == "Rural", :]

summary.head()
```

```python
#build the scatter plots for each city type
urban_plot=plt.scatter(urban["Total Riders"],urban["Avg. Fare"], marker="o", edgecolor="black",
           s=urban["driver_count"]*10, alpha=0.75, linewidth=1, c="lightcoral", label="Urban")

rural_plot=plt.scatter(rural["Total Riders"],rural["Avg. Fare"], marker="o", edgecolor="black",
           s=rural["driver_count"]*10, alpha=0.75, linewidth=1, c="gold", label="Rural")

suburban_plot=plt.scatter(suburban["Total Riders"],suburban["Avg. Fare"], marker="o", edgecolor="black",
           s=suburban["driver_count"]*10, alpha=0.75, linewidth=1, c="skyblue", label="Suburban")

#incorporate the other graph properties
plt.grid()

#set lower & upper limits for X-Axis
plt.xlim(0, 42)

#set lower & upper limits for Y-Axis
plt.ylim(18, 45)

#create a legend
legend = plt.legend(title="City Types", loc='best', prop={'size': 8})
legend.legendHandles[0]._sizes = [100]
legend.legendHandles[1]._sizes = [100]
legend.legendHandles[2]._sizes = [100]

#create labels for the scatter plot
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")

#incorporate a text label regarding circle size
textLabel = "Note: Circle size correlates with driver count per city"
plt.text(50, 40, textLabel, fontsize=12)

#save figure
plt.savefig("../pyber_bubble_plot.png")
```

```python
#calculate type percents
fare_percents = ride_city_data.groupby(["type"])
total_fares = fare_percents["fare"].sum()

#build pie chart
labels = ["Rural", "Suburban", "Urban"]

colors = ["gold", "skyblue", "lightcoral"]

explode = (0,0,.1)

plt.pie(total_fares, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)

plt.axis("equal")

plt.title("% of Total Fares by City Type")

#save figure
plt.savefig("../pyber_fares_by_city_type.png")
```

```python
#calculate type percents
ride_percents = ride_city_data.groupby(["type"])
total_rides = ride_percents["ride_id"].count()

#build pie chart
labels = ["Rural", "Suburban", "Urban"]

colors = ["gold", "skyblue", "lightcoral"]

explode = (0,0,.1)

plt.pie(total_rides, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)

plt.axis("equal")

plt.title("% of Total Rides by City Type")

#save figure
plt.savefig("../pyber_fares_by_city_type.png")
```

```python
#calculate type percents
driver_percents = city_data.groupby(["type"]).sum()
total_drivers = driver_percents["driver_count"] / city_data["driver_count"].sum()

#build pie chart
labels = ["Rural", "Suburban", "Urban"]

colors = ["gold", "skyblue", "lightcoral"]

explode = (0,0,.1)

plt.pie(total_drivers, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)

plt.axis("equal")

plt.title("% of Total Drivers by City Type")

#save figure
plt.savefig("../pyber_fares_by_city_type.png")
```
